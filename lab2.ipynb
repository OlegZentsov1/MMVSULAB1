#График населения Новосибирска

```
import matplotlib.pyplot as plt 
import pandas as pd
from scipy.integrate import odeint 


data = {'Y': [1991, 1992, 1993, 1994, 1995],
        'P': [1446000, 1442000, 1431000, 1418000, 1399000]}
df = pd.DataFrame(data)

plt.plot(df['Y'], df['P'])

plt.xlabel('Y')
plt.ylabel('P')
plt.title('Population of Novosibirsk')

plt.show()
```
