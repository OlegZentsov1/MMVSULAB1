```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def euler(f, x0, y0, h, n):
    x = [x0]
    y = [y0]
    for i in range(n - 1):
        xi = x[i]
        yi = y[i]
        yi1 = yi + h * dxdt(xi, yi)
        x.append(xi + h)  
        y.append(yi1)
    return x, y

def dxdt(x, y):
    k = 0.42
    return k * y

x0, y0 = 0, 1
h = 0.01
n = 10000
x = np.arange(0, 10 + h, h) 
t = np.linspace(0, 10, 1000)

x, y = euler(dxdt, x0 , y0, h, n = 1000)

solve = odeint(dxdt, y0 = 1, t = t)
solve = solve.flatten()

plt.plot(x, y, label = r"euler")
plt.plot(t, solve, label = r"$x' = k(x)*x$")
plt.grid(True)
plt.legend()
plt.show()
```
